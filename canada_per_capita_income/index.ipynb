```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_csv("canada_per_capita_income.csv")
df.head(3)
```

```python
%matplotlib inline
plt.xlabel('year')
plt.ylabel('per capita income (US$)')
plt.scatter(df.year, df.income, color='red', marker='+')
```

```python
reg = linear_model.LinearRegression()
reg.fit(df[['year']], df.income)
```

```python
# predict from 2017 to 2050
years = np.arange(1970, 2051)
```

```python
%matplotlib inline
plt.xlabel('year')
plt.ylabel('per capita income (US$)')
plt.scatter(df.year, df.income, color='red', marker='+')
plt.plot(years,reg.predict(years.reshape(-1,1)), color='blue')
```

```python
years = np.arange(2017, 2051)
p = reg.predict(years.reshape(-1,1))
df2 = pd.DataFrame(years, columns=['year'])
df2['income'] = p
df2.to_csv('canada_per_capita_income_predicted.csv', index=False)
```
